```python
import pandas as pd 
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
!pip install numpy seaborn random
```

```python
list = ["Food", "Travel", "Fashion", "Fitness", "Music", "Culture", "Family", "Health"]
```

```python
dates = pd.date_range('2021-01-01', periods=500)
category = [random.choice(list) for _ in range(500)]
likes = np.random.randint(0, 10000, size=500)

data_dictionary = {
    "Date":dates,
    "Category":category,
    "Like":likes
}
```

```python
data_dictionary
```

```python
df = pd.DataFrame(data_dictionary)
```

```python
category_counts = df['Category'].value_counts()
print(category_counts)
```

```python
df.dropna(inplace = True)
df.drop_duplicates(inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Like'] = df['Like'].apply(lambda x: int(x) if isinstance(x, str) and x.isdigit() else x)
```

```python
import seaborn as sns  # We're using the seaborn library for pretty plots
import matplotlib.pyplot as plt  # This is for showing our plots

sns.histplot(data=df, x='Like')  # This makes a histogram of 'Like'
plt.show()  # This shows the histogram on your screen
sns.boxplot(data=df, x='Category', y='Like')  # This makes a boxplot
plt.show()  # This shows the boxplot
# To find the average (mean) number of 'Like' overall
overall_mean_Like = df['Like'].mean()
print("Average Like Overall:", overall_mean_Like)

# To find the average (mean) number of 'Like' for each 'Category'
category_mean_Like = df.groupby('Category')['Like'].mean()
print("Average Like per Category:\n", category_mean_Like)
```
